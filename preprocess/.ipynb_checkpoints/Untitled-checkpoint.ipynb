## Split videos into tiles

```python
color_frames, n_frame = read_video('./videos/input.mp4')
mask_frames, n_frame = read_video('./videos/mask.mp4')
```

```python
tiles = {}
for j in range(0, 960, 240):
    for i in range(0, 960, 240):
         tiles[f'{j}_{i}'] = []
for frame in color_frames:
    for j in range(0, 960, 240):
        for i in range(0, 960, 240):
            tiles[f'{j}_{i}'].append(frame[j:j+240, i:i+240])
```

```python
mask_tiles = {}
for j in range(0, 960, 240):
    for i in range(0, 960, 240):
         mask_tiles[f'{j}_{i}'] = []
for frame in mask_frames:
    for j in range(0, 960, 240):
        for i in range(0, 960, 240):
            mask_tiles[f'{j}_{i}'].append(frame[j:j+240, i:i+240])
```

```python
fps = 20
resolution = 240
mkdir_if_missing('./videos/tiles')
for k in tqdm(tiles.keys()):
    save_path = f'./videos/tiles/input_{k}.mp4'
    save_video(save_path, tiles[k], fps, resolution=(resolution, resolution))
    save_path = f'./videos/tiles/mask_{k}.mp4'
    save_video(save_path, mask_tiles[k], fps, resolution=(resolution, resolution))
```

```python
xor_result = np.isclose(rgb_dest[0], rgb_gt, atol=0.1)  # each channel could have different results
mask = np.logical_or(xor_result[..., 2], np.logical_or(xor_result[..., 0], xor_result[..., 1]))
```

```python
mask = mask.astype(np.uint8)
mask = normalize_image_value(mask, convert_to='255')
```

```python
novel_view_to_save = normalize_image_value(rgb_dest[0], convert_to='255')
novel_view_gt_to_save = normalize_image_value(rgb_gt, convert_to='255')
cv2.imwrite('./images/1.png', novel_view_to_save[..., ::-1])
cv2.imwrite('./images/1_gt.png', novel_view_gt_to_save[..., ::-1])

# compute difference
ref1 = normalize_image_value(rgb_dest[0], convert_to='255')
ref2 = normalize_image_value(rgb_gt, convert_to='255')
difference = cv2.subtract(ref1, ref2)
difference = difference - difference.min()
difference = difference / difference.max()
difference *= 255
Conv_hsv_Gray = cv2.cvtColor(difference.astype(np.float32), cv2.COLOR_BGR2GRAY)
ret, mask = cv2.threshold(Conv_hsv_Gray.astype(np.uint8), 0, 255, cv2.THRESH_BINARY_INV |cv2.THRESH_OTSU)

cv2.imwrite('./images/1_mask.png', 255-mask) 
```

```python
mkdir_if_missing('./videos')
mesh_folder = '/home/tungi/dataset/mface/m--20180227--0000--6795937--GHS/tracked_mesh/E061_Lips_Puffed/'
tex_folder = '/home/tungi/dataset/mface/m--20180227--0000--6795937--GHS/unwrapped_uv_1024/E061_Lips_Puffed/average/'

# Create R and T of the source and target views
R_source, T_source = look_at_view_transform(400, 0, 0)
R_target, T_target = look_at_view_transform(400, 0, 15)

# Collect the frame numbers
frame_numbers = []
for p in Path(tex_folder).glob('*.png'):
    n = os.path.split(str(p))[-1].split('.')[0]
    frame_numbers.append(n)
frame_numbers.sort()

# Create the mesh of each frame
resolution = 1024
meshes = []
for n in tqdm(frame_numbers):
    obj, verts, tex = load_mf_obj(mesh_folder, tex_folder, n, tex_resolution=resolution)
    mesh = create_p3d_mesh(obj, verts, tex)
    meshes.append(mesh)

# Create the simulated rgbd video
z_far = 800
nview_frames = []
gt_frames = []
for mesh in tqdm(meshes):
    # novel view with holes
    rgbd_source = mesh2rgbd(mesh, R_source, T_source, resolution=resolution, zfar=z_far, return_cam=False)
    cloud = rgbd2cloud(rgbd_source, R_source, T_source, zfar=z_far)
    rgb_target = cloud2rgb(cloud, R_target, T_target)
    frame_to_save = normalize_image_value(rgb_target[0], convert_to='255')
    nview_frames.append(frame_to_save)
    # gt 
    rgbd_gt = mesh2rgbd(mesh, R_target, T_target, resolution=resolution, zfar=z_far, return_cam=False)
    rgb_gt = rgbd_gt[0][..., :3]
    frame_to_save = normalize_image_value(rgb_gt, convert_to='255')
    gt_frames.append(frame_to_save)

```
